```python
example = pd.read_csv('salary_example_submition.csv')
train = pd.read_csv('salary_train.csv')
test = pd.read_csv('salary_predict.csv')
train
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import normalize

ss = StandardScaler()

x = train.iloc[:, 1:7]
y = train.iloc[:, 7]
```
